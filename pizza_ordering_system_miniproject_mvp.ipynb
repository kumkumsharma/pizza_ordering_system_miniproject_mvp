```python
import os
import re
from datetime import datetime

# Create folders
os.makedirs("menu", exist_ok=True)
os.makedirs("logs", exist_ok=True)

# Create sample menu files

bases = """1,Thin Crust,120
2,Thick Crust,150
3,Cheese Burst,220
4,Whole Wheat,170
5,Multigrain,180
"""

pizzas = """1,Margherita,249
2,Farm House,349
3,Pepperoni Classic,399
4,Paneer Tikka,349
5,Veg Supreme,379
"""

toppings = """1,Extra Cheese,69
2,Jalapenos,49
3,Olives,59
4,Sweet Corn,39
5,Mushrooms,79
"""

with open("menu/bases.txt", "w") as f:
    f.write(bases)

with open("menu/pizzas.txt", "w") as f:
    f.write(pizzas)

with open("menu/toppings.txt", "w") as f:
    f.write(toppings)

print("Project folders created.")
print("Sample menu files created.")
```

```python
# load menu dynamically
def load_menu(file_path):
    """
    Load menu items from a text file.

    Expected file format:
    B1;Thin Crust;149

    Returns:
    [
        {
            "name": "Thin Crust",
            "price": 149.0
        },
        ...
    ]
    """

    menu = []

    # Check if file exists
    if not os.path.exists(file_path):
        print(f"ERROR: {file_path} not found.")
        return menu

    with open(file_path, "r", encoding="utf-8") as file:

        for line_no, line in enumerate(file, start=1):

            line = line.strip()

            # Ignore blank lines
            if line == "":
                continue

            parts = line.split(";")

            # Every line must have exactly 3 fields
            if len(parts) != 3:
                print(f"Skipping malformed line {line_no} in {file_path}")
                continue

            try:

                name = parts[1].strip()
                price = float(parts[2].strip())

                menu.append({
                    "name": name,
                    "price": price
                })

            except ValueError:
                print(f"Invalid price on line {line_no}")

    return menu
```

```python
def display_menu(menu, title):
    """
    Display menu with serial numbers.
    """

    print("\n" + "=" * 60)
    print(title.upper())
    print("=" * 60)

    print(f"{'No.':<6}{'Item Name':<35}{'Price (₹)':>10}")
    print("-" * 60)

    for index, item in enumerate(menu, start=1):

        print(
            f"{index:<6}"
            f"{item['name']:<35}"
            f"{item['price']:>10.2f}"
        )

    print("=" * 60)
```

```python
def validate_name(name):
    """
    Validate customer name.
    Rules:
    - Cannot be empty
    - Minimum 2 characters
    - Only alphabets and spaces
    """

    name = name.strip()

    if name == "":
        return False, "Name cannot be empty."

    if len(name) < 2:
        return False, "Name must contain at least 2 characters."

    if not re.fullmatch(r"[A-Za-z ]+", name):
        return False, "Name can contain only alphabets and spaces."

    return True, ""
```

```python
def validate_phone(phone):
    """
    Validate Indian mobile number.
    """

    phone = phone.strip()

    if not re.fullmatch(r"[6-9]\d{9}", phone):
        return (
            False,
            "Enter a valid 10-digit Indian mobile number starting with 6, 7, 8 or 9."
        )

    return True, phone
```

```python
def validate_quantity(quantity):
    """
    Validate quantity.
    Rules:
    - Integer only
    - Between 1 and 10
    """

    try:
        quantity = int(quantity)
    except ValueError:
        return False, "Quantity must be a whole number."

    if quantity < 1:
        return False, "Quantity cannot be less than 1."

    if quantity > 10:
        return False, "Maximum quantity allowed is 10."

    return True, quantity
```

```python
def validate_payment_mode(choice):
    """
    Validate payment mode.
    """

    payment_modes = {
        1: "Cash",
        2: "UPI",
        3: "Card"
    }

    try:
        choice = int(choice)
    except ValueError:
        return False, "Invalid payment mode."

    if choice not in payment_modes:
        return False, "Select 1, 2 or 3."

    return True, payment_modes[choice]
```

```python
def validate_menu_selection(choice, menu):
    """
    Validate menu selection.

    Returns:
    Selected menu item
    """

    try:
        choice = int(choice)

    except ValueError:
        return False, "Please enter a valid number."

    if choice < 1:
        return False, "Choice cannot be less than 1."

    if choice > len(menu):
        return False, "Choice is out of range."

    return True, menu[choice - 1]
```

```python
def get_menu_selection(menu, title):
    """
    Display menu and get a valid user selection.
    """

    while True:

        display_menu(menu, title)

        choice = input(f"\nSelect {title} (Enter Number): ")

        valid, result = validate_menu_selection(choice, menu)

        if valid:
            return result

        print(f"\n❌ {result}\n")
```

```python
def get_customer_details():
    """
    Get customer details from the user.

    Returns:
        {
            "name": customer_name,
            "phone": phone_number
        }
    """

    while True:

        name = input("Enter Customer Name : ")

        valid, message = validate_name(name)

        if valid:
            break

        print(f"\n❌ {message}\n")


    while True:

        phone = input("Enter Mobile Number : ")

        valid, message = validate_phone(phone)

        if valid:
            break

        print(f"\n❌ {message}\n")


    customer = {

        "name": name.strip(),

        "phone": phone.strip()

    }

    return customer
```

```python
def get_quantity():
    """
    Get pizza quantity from the user.

    Returns:
        int : Valid quantity
    """

    while True:

        quantity = input("Enter Quantity (1-10) : ")

        valid, result = validate_quantity(quantity)

        if valid:
            return result

        print(f"\n❌ {result}\n")
```

```python
def calculate_bill(base, pizza, toppings, quantity):
    """
    Calculate the complete bill.

    Business Rules:
    - Discount = 10% if quantity >= 5
    - GST = 18% after discount

    Returns:
        Dictionary containing complete bill details.
    """

    # Price of one pizza
    one_pizza_price = (
        base["price"]
        + pizza["price"]
        + toppings["price"]
    )

    # Before discount
    subtotal = one_pizza_price * quantity

    # Discount
    if quantity >= 5:
        discount = subtotal * 0.10
    else:
        discount = 0

    # Amount after discount
    taxable_amount = subtotal - discount

    # GST
    gst = taxable_amount * 0.18

    # Final Amount
    final_total = taxable_amount + gst

    bill = {

        "base_price": base["price"],

        "pizza_price": pizza["price"],

        "topping_price": toppings["price"],

        "quantity": quantity,

        "one_pizza_price": round(one_pizza_price, 2),

        "subtotal": round(subtotal, 2),

        "discount": round(discount, 2),

        "gst": round(gst, 2),

        "final_total": round(final_total, 2)

    }

    return bill
```

```python
def display_bill(customer,
                 base,
                 pizza,
                 topping,
                 bill):
    """
    Display the complete bill before payment.
    """

    print("\n")
    print("=" * 60)
    print("              PIZZA PARADISE")
    print("=" * 60)

    print(f"Customer Name : {customer['name']}")
    print(f"Mobile Number : {customer['phone']}")

    print("-" * 60)

    print(f"Base          : {base['name']}      ₹{base['price']:.2f}")
    print(f"Pizza         : {pizza['name']}      ₹{pizza['price']:.2f}")
    print(f"Topping       : {topping['name']}      ₹{topping['price']:.2f}")

    print(f"Quantity      : {bill['quantity']}")

    print("-" * 60)

    print(f"Price / Pizza : ₹{bill['one_pizza_price']:.2f}")
    print(f"Subtotal      : ₹{bill['subtotal']:.2f}")
    print(f"Discount      : ₹{bill['discount']:.2f}")
    print(f"GST (18%)     : ₹{bill['gst']:.2f}")

    print("-" * 60)

    print(f"TOTAL         : ₹{bill['final_total']:.2f}")

    print("=" * 60)
```

```python
def get_payment_mode():
    """
    Get payment mode from the user.

    Returns:
        str : Cash / UPI / Card
    """

    while True:

        print("\nSelect Payment Mode")
        print("-" * 25)
        print("1. Cash")
        print("2. UPI")
        print("3. Card")

        choice = input("\nEnter your choice (1-3): ")

        valid, result = validate_payment_mode(choice)

        if valid:
            return result

        print(f"\n❌ {result}\n")
```

```python
import json
import os
from datetime import datetime

def save_order(customer,
               base,
               pizza,
               topping,
               quantity,
               bill,
               payment_mode):
    """
    Save completed order to orders_log.txt

    Each order is stored as one JSON object,
    followed by a blank line.
    """

    order = {

        "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),

        "customer_name": customer["name"],

        "phone": customer["phone"],

        "base": base["name"],

        "pizza": pizza["name"],

        "topping": topping["name"],

        "base_price": base["price"],

        "pizza_price": pizza["price"],

        "topping_price": topping["price"],

        "quantity": quantity,

        "subtotal": bill["subtotal"],

        "discount": bill["discount"],

        "gst": bill["gst"],

        "final_total": bill["final_total"],

        "payment_mode": payment_mode

    }

    file_name = "orders_log.txt"

    try:

        with open(file_name, "a", encoding="utf-8") as file:

            json.dump(order, file, indent=4)

            file.write("\n\n")

        print("\n✅ Order saved successfully.")

    except Exception as e:

        print(f"\n❌ Error saving order : {e}")
```

```python
save_order(
    customer,
    selected_base,
    selected_pizza,
    selected_topping,
    quantity,
    bill,
    payment_mode
)
```

```python
def main():

    print("=" * 60)
    print("        WELCOME TO SLICEMATIC")
    print("=" * 60)

    # Load Menu
    bases = load_menu("menu/Types_of_Base.txt")
    pizzas = load_menu("menu/Types_of_Pizza.txt")
    toppings = load_menu("menu/Types_of_Toppings.txt")

    # Stop if menu loading fails
    if not bases or not pizzas or not toppings:
        print("\nUnable to load menu. Exiting application.")
        return

    # Customer Details
    customer = get_customer_details()

    # Menu Selection
    selected_base = get_menu_selection(bases, "Pizza Bases")

    selected_pizza = get_menu_selection(pizzas, "Pizza Types")

    selected_topping = get_menu_selection(toppings, "Pizza Toppings")

    # Quantity
    quantity = get_quantity()

    # Bill
    bill = calculate_bill(
        selected_base,
        selected_pizza,
        selected_topping,
        quantity
    )

    # Display Bill
    display_bill(
        customer,
        selected_base,
        selected_pizza,
        selected_topping,
        bill
    )

    # Payment
    payment_mode = get_payment_mode()

    print(f"\nPayment Successful using {payment_mode}")

    # Save Order
    save_order(
        customer,
        selected_base,
        selected_pizza,
        selected_topping,
        quantity,
        bill,
        payment_mode
    )

    print("\nThank you for ordering with SliceMatic!")
```

```python
main()
```
